```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv')
```

```python
linhas, colunas = df.shape[0], df.shape[1]
```

```python
df = df.rename(columns={
    'work_year': 'Ano',
    'experience_level': 'Senioridade',
    'employment_type': 'Contrato',
    'job_title': 'Cargo',
    'salary': 'Salário',
})
display(df.head())
```

```python
df = df.rename(columns={
    'salary_currency': 'Moeda',
    'salary_in_usd': 'USD',
    'employee_residence': 'Residência',
    'remote_ratio': 'Remoto',
    'company_location': 'empresa',
    'company_size': 'Tamanho da empresa',
})
display(df.head(10))
```

```python
df['Senioridade'].value_counts()
```

```python
df['Senioridade'] = df['Senioridade'].map({
    'MI': 'Pleno',
    'SE': 'Sênior',
    'EN': 'Júnior',
    'EX': 'Executivo',
})
```

```python
df['Contrato'].value_counts()
```

```python
df['Contrato'] = df['Contrato'].map({
    'FT': 'Tempo Integral',
    'PT': 'Meio Integral',
    'FL': 'Flexível',
    'CT': 'Contrato temporário'
})
```

```python
df['Tamanho da empresa'].value_counts()
```

```python
df['Tamanho da empresa'] = df['Tamanho da empresa'].map({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})
```

```python
df['Remoto'].value_counts()
```

```python
df['Remoto'] = df['Remoto'].map({
    0: 'Presencial',
    50: 'Hibrido',
    100: 'Remoto'
})
```

```python
df.head()
```

```python
df.head(10)
```

```python
df['Salário_Mensal'] =df['Salário'] / 12
```

```python
df['Salário_Mensal'] = df['Salário_Mensal'].round(2)
```
